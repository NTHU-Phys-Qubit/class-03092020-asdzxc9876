```python
import matplotlib.pyplot as plt
import numpy as np
```

## Problem 1

Plot sin(x)

```python
def sine(x):
    y = np.sin(x)
    return y

x = np.linspace(0, 10, 1000)
y = sine(x)

fig, ax = plt.subplots(figsize=(20, 7))

ax.plot(x, y , color='blue', linewidth=2.2, linestyle='-', label = 'sin function')
ax.legend()
ax.set_xlabel('Time (ms)')
ax.set_ylabel('value')
ax.set_title('Sin plot')

plt.grid()
```

## Problem 2

Pefrom some matrix multiplication 

```python
def multiplication(A, B):
    row = np.array(A.shape)[0]
    column =  np.array(A.shape)[1]
    C = np.zeros((row, column))
    for i in range(row):
        for j in range(column):
            summation = 0
            for k in range(np.array(A.shape)[1]):
                summation += A[i, k] * B[k, j]
            C[i, j] = summation
    return C
    
###   test banch   ###

m1 = np.array([[10,13],[5,1],[8,9]])
m2 = np.array([[10,6,7],[5,3,2]])

result = multiplication(m1, m2)
print('Matrix1 : ', m1)
print('Matrix2 : ', m2)
print('Multiplication result : ', result)
```
